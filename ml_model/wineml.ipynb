```python
# Imports
import pandas as pd
import numpy as np
import joblib

from sklearn.model_selection import train_test_split
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.ensemble import RandomForestRegressor
```

```python
# Loading the data
df_wine = pd.read_csv('data.csv')
df_wine
```

```python
# Inspecting the data
df_wine.info()
```

```python
df_wine.describe()
```

```python
# Splitting the train and test data
train_set, test_set = train_test_split(df_wine, test_size=0.2, random_state=42)
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)

# For loop
for train_index, test_index in split.split(df_wine, df_wine['Alchohol']):
  stratified_train_set = df_wine.loc[train_index]
  stratified_test_set = df_wine.loc[test_index]

# Creating the dataset
wine_df = stratified_train_set.copy()
wine_df = stratified_train_set.drop('Wine', axis=1)
wine_df_labels = stratified_train_set['Wine'].copy()
```

```python
# Building the model
model = RandomForestRegressor()
model.fit(wine_df, wine_df_labels)
print("Model Training Ends...")
test_features=stratified_test_set.drop("Wine", axis=1)
test_labels=stratified_test_set["Wine"].copy()
y_labels=model.predict(test_features)
x=list(y_labels)
y=list(test_labels)
accuracy=[]

# Check for the accuracy
for i in range(len(test_labels)):
	if x[i]>y[i]:
		accuracy.append((y[i]/x[i])*100)
	else:
		accuracy.append((x[i]/y[i])*100)
joblib.dump(model, "rf_model.joblib")
print("Model Saved...")
acc=sum(accuracy)/len(x)
print("Final Accuracy of the Model: ", acc)
```
